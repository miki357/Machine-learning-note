## Merge, Join, Concatenate

```python
Image(filename = "img/merge.png", width = 800, height = 400)
```

### Merge

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})  
right
```

### * inner join 只取交集

```python
# 合併left, right, 方式為inner, 以'key'判斷
pd.merge(left,right,how='inner',on='key')   
```

### Join

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2']) 
left
```

```python
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
right
```

```python
left.join(right)
```

### * outer join 取全部檔案 會有空值!

```python
left.join(right, how='outer')
```

### Concatenate

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7]) 
df2
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
df3
```

```python
pd.concat([df1,df2,df3])   #欄位一樣 增加row資料
```

```python
pd.concat([df1,df2,df3],axis=1)    #axis預設為0代表row, 1代表column
```

### Script
- Requests 自動下載網頁 
- beautifulsoup 解析原始碼

```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR3QwTlRFU0JYcG9MVlJYS0FBUAE?hl=zh-TW&gl=TW&ceid=TW%3Azh-Hant'
# url 只是變數名稱,但通常是網址的意思

r = requests.get(url)
web_content = r.text
soup = BeautifulSoup(web_content, 'lxml')
```

#### 檢視新聞標題在HTML的規則發現：
- 每個文章都被包覆在class = ‘xrnccd’的div物件中
- 整個文章點擊區塊連結: class = “VDXfz”
- 文章中的標題則是包覆在class為’h3’的物件中

連結 https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR3QwTlRFU0JYcG9MVlJYS0FBUAE?hl=zh-TW&gl=TW&ceid=TW%3Azh-Hant

```python
# 找出所有標題區塊的元素並儲存為一個list
title = soup.find_all('div', class_='xrnccd')
print(title)
```

```python
# 從標題區塊list中, 找出<h3>...</h3>標籤內的標題文字
titles = [t.find('h3').text for t in title]
titles
```

```python
# 找出新聞標題所對應的新聞文章資料來源
newUrls = [requests.get(t.find('a')['href'].replace('.','https://news.google.com',1)).url for t in title] 
newUrls
```

```python
df = pd.DataFrame({'title': titles,'links': newUrls})
df
```

## Supplement
取出新聞全文

```python
url = df['links'][4]
print(url)
r = requests.get(url)
web_content = r.text
soup = BeautifulSoup(web_content,'lxml')
articleContent = soup.find_all('p')
articleContent
# 將所有tag p中的text取出並放進list。
article = []
for p in articleContent:
    article.append(p.text)
    articleAll = '\n'.join(article)
print(articleAll)
```
